### **Fetch Data**

In previous notebook decided that 2400 - 2700 retention time was an appropriate range. 

```python
expOut = MSExperiment()
```

```python
rtLow = 2400
rtHigh = 2700
for s in exp.getSpectra():
    rt = s.getRT()
    if (rt < rtHigh) and (rt > rtLow):
        expOut.addSpectrum(s)
```

```python
expOut.getNrSpectra()
```

#### **Create tsv data**

Since the `.osw` file is hard to interpret and the `.tsv` does not contain IM information, create a .tsv file for Ira with the IM information. Note that this will take a while to run. 

I processed the .osw to a .tsv using the `exportPyprophetOutput.py` script so now I just want to filter to the desired RTT range. 

```python
oswRslts = pd.read_csv("run_1330_0_pyprophet_output.tsv", sep="\t")
```

```python
oswRsltsFiltered = oswRslts[
    (oswRslts["RT"].between(rtLow, rtHigh)) & (oswRslts["mz"].between(700, 725))
]
```
